## CIFAR-10 Autoencoder Solutions
### Question 1: Denoising Autoencoder
### Question 2: Variational Autoencoder (VAE) for Image Reconstruction

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt

print("TensorFlow version:", tf.__version__)
print("=" * 80)

# ==========================================
# PART 1: DENOISING AUTOENCODER
# ==========================================

class DenoisingAutoencoder:
    """Denoising Autoencoder for image cleaning"""

    def __init__(self):
        self.model = None
        self.history = None

    @staticmethod
    def add_noise(images, noise_factor=0.15):
        """Add Gaussian noise to images"""
        noisy_images = images + noise_factor * np.random.normal(
            loc=0.0, scale=1.0, size=images.shape
        )
        return np.clip(noisy_images, 0., 1.)

    def build_model(self):
        """Build the denoising autoencoder architecture"""
        input_img = layers.Input(shape=(32, 32, 3))

        # Encoder
        x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
        x = layers.MaxPooling2D((2, 2), padding='same')(x)
        x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
        encoded = layers.MaxPooling2D((2, 2), padding='same')(x)

        # Decoder
        x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(encoded)
        x = layers.UpSampling2D((2, 2))(x)
        x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
        x = layers.UpSampling2D((2, 2))(x)
        decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

        self.model = models.Model(input_img, decoded, name='denoising_autoencoder')
        self.model.compile(optimizer='adam', loss='binary_crossentropy')

        return self.model

    def train(self, x_train, x_test, noise_factor=0.15, epochs=10, batch_size=128):
        """Train the denoising autoencoder"""
        # Create noisy versions
        x_train_noisy = self.add_noise(x_train, noise_factor)
        x_test_noisy = self.add_noise(x_test, noise_factor)

        print("\n" + "=" * 80)
        print("TRAINING DENOISING AUTOENCODER")
        print("=" * 80)

        self.history = self.model.fit(
            x_train_noisy, x_train,
            epochs=epochs,
            batch_size=batch_size,
            shuffle=True,
            validation_data=(x_test_noisy, x_test),
            verbose=1
        )

        return x_test_noisy

    def visualize_results(self, x_test, x_test_noisy, n=10):
        """Visualize denoising results"""
        decoded_imgs = self.model.predict(x_test_noisy[:n], verbose=0)

        plt.figure(figsize=(20, 6))
        plt.suptitle('Denoising Autoencoder: Original → Noisy → Denoised',
                     fontsize=16, fontweight='bold')

        for i in range(n):
            # Original
            ax = plt.subplot(3, n, i + 1)
            plt.imshow(x_test[i])
            if i == 0:
                plt.ylabel("Original", fontsize=12, fontweight='bold')
            plt.axis("off")

            # Noisy
            ax = plt.subplot(3, n, i + 1 + n)
            plt.imshow(x_test_noisy[i])
            if i == 0:
                plt.ylabel("Noisy Input", fontsize=12, fontweight='bold')
            plt.axis("off")

            # Denoised
            ax = plt.subplot(3, n, i + 1 + 2 * n)
            plt.imshow(np.clip(decoded_imgs[i], 0, 1))
            if i == 0:
                plt.ylabel("Denoised", fontsize=12, fontweight='bold')
            plt.axis("off")

        plt.tight_layout()
        plt.show()

    def plot_training_history(self):
        """Plot training loss curves"""
        if self.history is None:
            print("No training history available!")
            return

        plt.figure(figsize=(10, 4))
        plt.plot(self.history.history['loss'], label='Training Loss', linewidth=2)
        plt.plot(self.history.history['val_loss'], label='Validation Loss', linewidth=2)
        plt.title('Denoising Autoencoder Training History', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Loss', fontsize=12)
        plt.legend(fontsize=11)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()


# ==========================================
# PART 2: VARIATIONAL AUTOENCODER (VAE)
# ==========================================

class Sampling(layers.Layer):
    """Sampling layer for VAE: samples from N(z_mean, exp(z_log_var))"""

    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon


class VAE(keras.Model):
    """Variational Autoencoder with custom training loop"""

    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker
        ]

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Reconstruction loss
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(
                        tf.reshape(data, [-1, 32 * 32 * 3]),
                        tf.reshape(reconstruction, [-1, 32 * 32 * 3])
                    )
                )
            )

            # KL divergence loss
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))

            total_loss = reconstruction_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

    def test_step(self, data):
        if isinstance(data, tuple):
            data = data[0]

        z_mean, z_log_var, z = self.encoder(data)
        reconstruction = self.decoder(z)

        reconstruction_loss = tf.reduce_mean(
            tf.reduce_sum(
                keras.losses.binary_crossentropy(data, reconstruction)
            )
        )

        kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
        kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))

        total_loss = reconstruction_loss + kl_loss

        return {
            "loss": total_loss,
            "reconstruction_loss": reconstruction_loss,
            "kl_loss": kl_loss,
        }

    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        return self.decoder(z)


class VariationalAutoencoder:
    """Wrapper class for VAE model"""

    def __init__(self, latent_dim=32):
        self.latent_dim = latent_dim
        self.encoder = None
        self.decoder = None
        self.vae = None
        self.history = None

    def build_encoder(self):
        """Build encoder network"""
        encoder_inputs = keras.Input(shape=(32, 32, 3))
        x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
        x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
        x = layers.Flatten()(x)
        x = layers.Dense(16, activation="relu")(x)

        z_mean = layers.Dense(self.latent_dim, name="z_mean")(x)
        z_log_var = layers.Dense(self.latent_dim, name="z_log_var")(x)
        z = Sampling()([z_mean, z_log_var])

        self.encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
        return self.encoder

    def build_decoder(self):
        """Build decoder network"""
        latent_inputs = keras.Input(shape=(self.latent_dim,))
        x = layers.Dense(8 * 8 * 64, activation="relu")(latent_inputs)
        x = layers.Reshape((8, 8, 64))(x)
        x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
        x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
        decoder_outputs = layers.Conv2DTranspose(3, 3, activation="sigmoid", padding="same")(x)

        self.decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
        return self.decoder

    def build_model(self):
        """Build complete VAE model"""
        self.build_encoder()
        self.build_decoder()
        self.vae = VAE(self.encoder, self.decoder)
        self.vae.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001))
        return self.vae

    def train(self, x_train, x_test, epochs=15, batch_size=128):
        """Train the VAE"""
        print("\n" + "=" * 80)
        print("TRAINING VARIATIONAL AUTOENCODER (VAE)")
        print("=" * 80)

        self.history = self.vae.fit(
            x_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(x_test, x_test),
            verbose=1
        )

    def visualize_results(self, x_test, n=10):
        """Visualize VAE reconstruction results"""
        z_mean, z_log_var, z = self.encoder.predict(x_test[:n], verbose=0)
        reconstructed = self.decoder.predict(z, verbose=0)

        plt.figure(figsize=(20, 4))
        plt.suptitle('VAE Reconstruction: Original (top) vs Reconstructed (bottom)',
                     fontsize=16, fontweight='bold')

        for i in range(n):
            # Original
            ax = plt.subplot(2, n, i + 1)
            plt.imshow(x_test[i])
            if i == 0:
                plt.ylabel("Original", fontsize=12, fontweight='bold')
            plt.axis("off")

            # Reconstructed
            ax = plt.subplot(2, n, i + 1 + n)
            plt.imshow(np.clip(reconstructed[i], 0, 1))
            if i == 0:
                plt.ylabel("Reconstructed", fontsize=12, fontweight='bold')
            plt.axis("off")

        plt.tight_layout()
        plt.show()

    def plot_training_history(self):
        """Plot VAE training metrics"""
        if self.history is None:
            print("No training history available!")
            return

        plt.figure(figsize=(15, 4))

        # Total Loss
        plt.subplot(1, 3, 1)
        plt.plot(self.history.history['loss'], label='Training', linewidth=2)
        if 'val_loss' in self.history.history:
            plt.plot(self.history.history['val_loss'], label='Validation', linewidth=2)
        plt.title('Total Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Loss', fontsize=12)
        plt.legend(fontsize=10)
        plt.grid(True, alpha=0.3)

        # Reconstruction Loss
        plt.subplot(1, 3, 2)
        plt.plot(self.history.history['reconstruction_loss'], label='Training', linewidth=2)
        if 'val_reconstruction_loss' in self.history.history:
            plt.plot(self.history.history['val_reconstruction_loss'], label='Validation', linewidth=2)
        plt.title('Reconstruction Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Loss', fontsize=12)
        plt.legend(fontsize=10)
        plt.grid(True, alpha=0.3)

        # KL Loss
        plt.subplot(1, 3, 3)
        plt.plot(self.history.history['kl_loss'], label='Training', linewidth=2)
        if 'val_kl_loss' in self.history.history:
            plt.plot(self.history.history['val_kl_loss'], label='Validation', linewidth=2)
        plt.title('KL Divergence Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=12)
        plt.ylabel('Loss', fontsize=12)
        plt.legend(fontsize=10)
        plt.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()


# ==========================================
# MAIN EXECUTION
# ==========================================

def main():
    """Main execution function"""

    # Load and preprocess CIFAR-10 dataset
    print("Loading CIFAR-10 dataset...")
    (x_train, _), (x_test, _) = cifar10.load_data()
    x_train = x_train.astype("float32") / 255.0
    x_test = x_test.astype("float32") / 255.0

    print(f"Training data shape: {x_train.shape}")
    print(f"Test data shape: {x_test.shape}")

    # ==========================================
    # Part 1: Denoising Autoencoder
    # ==========================================
    print("\n" + "=" * 80)
    print("PART 1: DENOISING AUTOENCODER")
    print("=" * 80)

    denoiser = DenoisingAutoencoder()
    denoiser.build_model()
    print("\nDenoising Autoencoder Model Summary:")
    denoiser.model.summary()

    # Train denoising autoencoder
    x_test_noisy = denoiser.train(
        x_train, x_test,
        noise_factor=0.15,
        epochs=10,
        batch_size=128
    )

    # Visualize denoising results
    print("\nGenerating denoising visualizations...")
    denoiser.visualize_results(x_test, x_test_noisy, n=10)
    denoiser.plot_training_history()

    # ==========================================
    # Part 2: Variational Autoencoder (VAE)
    # ==========================================
    print("\n" + "=" * 80)
    print("PART 2: VARIATIONAL AUTOENCODER (VAE)")
    print("=" * 80)

    vae_model = VariationalAutoencoder(latent_dim=32)
    vae_model.build_model()

    print("\nVAE Encoder Summary:")
    vae_model.encoder.summary()
    print("\nVAE Decoder Summary:")
    vae_model.decoder.summary()

    # Train VAE
    vae_model.train(x_train, x_test, epochs=15, batch_size=128)

    # Visualize VAE results
    print("\nGenerating VAE visualizations...")
    vae_model.visualize_results(x_test, n=10)
    vae_model.plot_training_history()

    # ==========================================
    # Final Summary
    # ==========================================
    print("\n" + "=" * 80)
    print("TRAINING COMPLETE!")
    print("=" * 80)
    print("✓ Denoising Autoencoder: Trained and visualized")
    print("✓ Variational Autoencoder: Trained and visualized")
    print("\nBoth models have been successfully trained on CIFAR-10 dataset!")
    print("=" * 80)


if __name__ == "__main__":
    main()
```
